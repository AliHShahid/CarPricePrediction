```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
mpl.style.use('ggplot')
```

```python
car=pd.read_csv('car_price_prediction.csv')
```

```python
car.head()
```

```python

import seaborn as sns
```

```python

plt.subplots(figsize=(15,7))
ax=sns.boxplot(x='Model',y='Price',data=car)
ax.set_xticklabels(ax.get_xticklabels(),rotation=40,ha='right')
plt.show()
```

```python
plt.subplots(figsize=(20,10))
ax=sns.swarmplot(x='Prod. year',y='Price',data=car)
ax.set_xticklabels(ax.get_xticklabels(),rotation=40,ha='right')
plt.show()
```

```python
sns.relplot(x='Mileage',y='Price',data=car,height=7,aspect=1.5)
```
